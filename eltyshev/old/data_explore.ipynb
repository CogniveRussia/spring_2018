```python
df = pd.read_csv("../data/off_members.csv")
```

```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
    
def encode_non_numbers(df, col):
    df.loc[df[col].apply(is_number) == False, col] = df.loc[df[col].apply(is_number) == False, col].factorize()[0] * (-1) - 1
    df[col] = df[col].astype(float)

encode_non_numbers(df, "P_BSCLIENTID")
```
