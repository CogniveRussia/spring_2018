```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

pd.options.display.float_format = '{:.0f}'.format
```

```python
members = pd.read_csv("../data/off_members.csv")
ops = pd.read_csv("../data/off_ops.csv")
```

```python
merged = members[["P_CLIENTID", "P_OPERATIONID"]].merge(ops[["ID", "P_EKNPCODE", "P_OPERATIONDATETIME"]], left_on="P_OPERATIONID", right_on="ID")
```

```python
null_mask = merged.P_EKNPCODE.isnull() | merged.P_CLIENTID.isnull()
corrupted_data_mask = merged.P_EKNPCODE.isin(["З46"])
mask = null_mask | corrupted_data_mask
merged_not_null = merged.loc[~mask].sort_values(by=["P_CLIENTID", "P_OPERATIONDATETIME"])

merged_not_null["P_EKNPCODE"] = merged_not_null["P_EKNPCODE"].astype(int)

agged = merged_not_null[["P_CLIENTID", "P_EKNPCODE"]].groupby("P_CLIENTID").agg(
    lambda x: ",".join(map(lambda x: str(int(x)), x.values))).reset_index()
```

```python
agged.to_csv("data/client_knp.csv", index=False)
```
